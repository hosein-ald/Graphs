```python
import numpy as np
import matplotlib.pyplot as plt

# Synthetic survival data
np.random.seed(3)
# Survival times in days
control_times = np.random.exponential(scale=10, size=50)  
treatment_times = np.random.exponential(scale=15, size=50)

# Sort times
control_times.sort()
treatment_times.sort()

def compute_km_curve(times):
    """
    Returns time points (t) and survival probabilities (S).
    """
    n = len(times)
    # Each event reduces survival fraction
    t = []
    S = []
    surv_frac = 1.0
    for i, time in enumerate(times):
        # At each time, an event (death) occurs
        surv_frac *= (n - (i + 1))/ (n - i)  # Adjust for each event
        t.append(time)
        S.append(surv_frac)
    return np.array(t), np.array(S)

t_control, s_control = compute_km_curve(control_times)
t_treatment, s_treatment = compute_km_curve(treatment_times)

plt.figure()

plt.step(t_control, s_control, where='post', label='Control')
plt.step(t_treatment, s_treatment, where='post', label='Treatment')

plt.xlabel('Time (days)')
plt.ylabel('Survival Probability')
plt.title('Kaplan–Meier Survival Curves')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(5)
num_genes = 100
log2_fold_change = np.random.normal(0, 2, num_genes)  # random log2 fold changes
p_values = np.random.uniform(0, 1, num_genes)         # random p-values
neg_log10_p_values = -np.log10(p_values)

plt.figure()

plt.scatter(log2_fold_change, neg_log10_p_values)
plt.axhline(y=-np.log10(0.05), linestyle='--')  # significance threshold
plt.axvline(x=1.0, linestyle='--')              # up-regulation threshold
plt.axvline(x=-1.0, linestyle='--')             # down-regulation threshold

plt.xlabel('Log2 Fold Change')
plt.ylabel('-log10(p-value)')
plt.title('Volcano Plot: Synthetic Gene Data')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

np.random.seed(6)
# Synthetic data: 3 clusters in 4D space
cluster1 = np.random.normal(loc=0, scale=1, size=(50,4))
cluster2 = np.random.normal(loc=5, scale=1, size=(50,4))
cluster3 = np.random.normal(loc=10, scale=1, size=(50,4))

X = np.vstack([cluster1, cluster2, cluster3])  # Combine

# Perform PCA to reduce to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure()

plt.scatter(X_pca[:,0], X_pca[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Plot: Synthetic 4D Data -> 2D')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)
n = 30
method1 = np.random.normal(10, 1, n)
method2 = method1 + np.random.normal(0, 0.5, n)  # second method slightly differs

mean_measurement = (method1 + method2) / 2
diff_measurement = method1 - method2

plt.figure()

plt.scatter(mean_measurement, diff_measurement)
plt.axhline(np.mean(diff_measurement), linestyle='--')
plt.axhline(np.mean(diff_measurement) + 1.96*np.std(diff_measurement), linestyle='--')
plt.axhline(np.mean(diff_measurement) - 1.96*np.std(diff_measurement), linestyle='--')

plt.xlabel('Mean of Two Methods')
plt.ylabel('Difference (Method1 - Method2)')
plt.title('Bland–Altman Plot')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(15)
num_points = 200
chromosomes = np.random.randint(1, 6, num_points)  # chromosome IDs from 1 to 5
positions = np.random.randint(1e3, 1e5, num_points) 
pvals = np.random.rand(num_points)
neg_log10_pvals = -np.log10(pvals)

# Sort by chromosome then by position
sorted_idx = np.lexsort((positions, chromosomes))
chromosomes = chromosomes[sorted_idx]
positions = positions[sorted_idx]
neg_log10_pvals = neg_log10_pvals[sorted_idx]

# Create a cumulative position to place them in one dimension
unique_chroms = np.unique(chromosomes)
current_pos = 0
x_values = []
chr_offsets = {}

for chrom in unique_chroms:
    chrom_idx = np.where(chromosomes == chrom)[0]
    start_pos = positions[chrom_idx[0]]
    # SHIFT positions so each chromosome starts after the previous
    for idx in chrom_idx:
        x_values.append(current_pos + (positions[idx] - start_pos))
    current_pos += (positions[chrom_idx[-1]] - start_pos) + 1000  # gap after each chromosome
    chr_offsets[chrom] = current_pos

x_values = np.array(x_values)

plt.figure()

plt.scatter(x_values, neg_log10_pvals, s=10)
plt.xlabel('Chromosome')
plt.ylabel('-log10(p-value)')
plt.title('Manhattan Plot (Synthetic)')

plt.tight_layout()
plt.show()
```
