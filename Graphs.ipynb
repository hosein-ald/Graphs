```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
genes = ['GeneA', 'GeneB', 'GeneC', 'GeneD', 'GeneE', 'GeneF']
wt_expression = [10, 15, 8, 9, 3, 3]   # Wild-type
mut_expression = [8, 20, 12, 3, 6, 7]  # Mutant

# Create x-coordinates for bars
x = np.arange(len(genes))

# Bar width
width = 0.4

plt.figure()  # Each plot in its own figure

# Plot bars for wild-type and mutant side by side
plt.bar(x - width/2, wt_expression, width, label='Wild-type')
plt.bar(x + width/2, mut_expression, width, label='Mutant')

# Labeling
plt.xticks(x, genes)
plt.xlabel('Gene')
plt.ylabel('Expression Level')
plt.title('Bar Plot: Gene Expression in Wild-type vs. Mutant')
plt.legend()

plt.tight_layout()
plt.show()
```
