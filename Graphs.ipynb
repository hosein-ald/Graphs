```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
genes = ['GeneA', 'GeneB', 'GeneC']
wt_expression = [10, 15, 8]   # Wild-type
mut_expression = [8, 20, 12]  # Mutant

# Create x-coordinates for bars
x = np.arange(len(genes))

# Bar width
width = 0.4

plt.figure()  # Each plot in its own figure

# Plot bars for wild-type and mutant side by side
plt.bar(x - width/2, wt_expression, width, label='Wild-type')
plt.bar(x + width/2, mut_expression, width, label='Mutant')

# Labeling
plt.xticks(x, genes)
plt.xlabel('Gene')
plt.ylabel('Expression Level')
plt.title('Bar Plot: Gene Expression in Wild-type vs. Mutant')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Synthetic data
np.random.seed(0)
control_cell_sizes = np.random.normal(loc=10, scale=2, size=50)  # Control group
treated_cell_sizes = np.random.normal(loc=12, scale=2.5, size=50) # Treated group

# Combine into one list for boxplot
data_to_plot = [control_cell_sizes, treated_cell_sizes]

plt.figure()

# Boxplot: positions 1 and 2 for the two groups
plt.boxplot(data_to_plot, positions=[1, 2], widths=0.6)

# Labeling
plt.xticks([1, 2], ['Control', 'Treated'])
plt.ylabel('Cell Size (m)')
plt.title('Box Plot: Cell Size Distributions')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
group1 = np.random.normal(5, 1.5, 100)
group2 = np.random.normal(6, 2, 100)
group3 = np.random.normal(7, 1, 100)

all_groups = [group1, group2, group3]

plt.figure()

# Violin plots
violin_parts = plt.violinplot(all_groups, positions=[1,2,3], showmeans=True, showextrema=True)

plt.xticks([1,2,3], ['Group1', 'Group2', 'Group3'])
plt.ylabel('Gene Expression Level')
plt.title('Violin Plot: Gene Expression in 3 Groups')

plt.tight_layout()
plt.show()
```
